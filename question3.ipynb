```python
# Calculating Information Gain before the 2 new rows are added

# 2 Targets --> play - yes, no
# 4 Attributes - outlook, temp, humidity, wind

# -----------------------------------------------------------------------------
# Attribute 1
# outlook - sunny, overcast, rain

# Sunny Entropy - 3N, 2Y
sunnyEnt0 = (3/5)*math.log(5/3) + (2/5)*math.log(5/2)

# Overcast Entropy - 4Y, 0N
overcastEnt0 = 0

# Rain Entropy - 3Y, 2N
rainEnt0 = (3/5)*math.log(5/3) + (2/5)*math.log(5/2)

# Info Gain = 1 - AvgEntropy
outlookIG0 = 1 - (sunnyEnt0*(5/14) + overcastEnt0*(4/14) + rainEnt0*(5/14))
print("Outlook Info Gain: ", outlookIG0)




# -----------------------------------------------------------------------------
# Attribute 2
# temp - hot, mild, cool

# Hot Entropy - 2N, 2Y
hotEnt0 = 1

# Mild Entropy - 4Y, 2N
mildEnt0 = (4/6)*math.log(6/4) + (2/6)*math.log(6/2)

# Cool Entropy - 3Y, 1N
coolEnt0 = (3/4)*math.log(4/3) + (1/4)*math.log(4/1)

# Info Gain = 1 - AvgEntropy
tempIG0 = 1 - (hotEnt0*(4/14) + mildEnt0*(6/14) + coolEnt0*(4/14))
print("Temp Info Gain: ", tempIG0)


# -----------------------------------------------------------------------------
# Attribute 3
# humidity - high, normal

# High Entropy - 4N, 3Y
highEnt0 = (4/7)*math.log(7/4) + (3/7)*math.log(7/3)

# Normal Entropy - 6Y, 1N
normalEnt0 = (6/7)*math.log(7/6) + (1/6)*math.log(6/1)


# Info Gain = 1 - AvgEntropy
humidityIG0 = 1 - (highEnt0*(7/14) + normalEnt0*(7/14))
print("Humidity Info Gain: ", humidityIG0)


# -----------------------------------------------------------------------------
# Attribute 4
# wind - weak, strong

# Weak Entropy - 6Y, 2N
weakEnt0 = (6/8)*math.log(8/6) + (2/8)*math.log(8/2)

# Strong Entropy - 3N, 3Y
strongEnt0 = 1


# Info Gain = 1 - AvgEntropy
windIG0 = 1 - (weakEnt0*(8/14) + strongEnt0*(6/14))
print("Wind Info Gain: ", windIG0)


# Which Atrribute to choose at the root? We want the largest information gain
infoGains0 = [outlookIG0, tempIG0, humidityIG0, windIG0]
infoGains0.sort()

print("The largest information gain is ", infoGains0[-1])

print("The attribute chosen at the root is Outlook, since it has the largest information gain.")
```

```python
# Calculating Information Gain after the 2 new rows are added

# 2 Targets --> play - yes, no
# 4 Attributes - outlook, temp, humidity, wind

# -----------------------------------------------------------------------------
# Attribute 1
# outlook - sunny, overcast, rain

# Sunny Entropy - 3N, 2Y
sunnyEnt = (3/5)*math.log(5/3) + (2/5)*math.log(5/2)

# Overcast Entropy - 4Y, 2N
overcastEnt = (4/6)*math.log(6/4) + (2/6)*math.log(6/2)

# Rain Entropy - 3Y, 2N
rainEnt = (3/5)*math.log(5/3) + (2/5)*math.log(5/2)

# Info Gain = 1 - AvgEntropy
outlookIG = 1 - (sunnyEnt*(5/16) + overcastEnt*(6/16) + rainEnt*(5/16))
print("Outlook Info Gain: ", outlookIG)




# -----------------------------------------------------------------------------
# Attribute 2
# temp - hot, mild, cool

# Hot Entropy - 2N, 2Y
hotEnt = 1

# Mild Entropy - 4Y, 4N
mildEnt = 1

# Cool Entropy - 3Y, 1N
coolEnt = (3/4)*math.log(4/3) + (1/4)*math.log(4/1)

# Info Gain = 1 - AvgEntropy
tempIG = 1 - (hotEnt*(4/16) + mildEnt*(8/16) + coolEnt*(4/16))
print("Temp Info Gain: ", tempIG)


# -----------------------------------------------------------------------------
# Attribute 3
# humidity - high, normal

# High Entropy - 4N, 3Y
highEnt = (4/7)*math.log(7/4) + (3/7)*math.log(7/3)

# Normal Entropy - 6Y, 3N
normalEnt = (6/9)*math.log(9/6) + (3/9)*math.log(9/3)


# Info Gain = 1 - AvgEntropy
humidityIG = 1 - (highEnt*(7/16) + normalEnt*(9/16))
print("Humidity Info Gain: ", humidityIG)


# -----------------------------------------------------------------------------
# Attribute 4
# wind - weak, strong

# Weak Entropy - 6Y, 3N
weakEnt = (6/9)*math.log(9/6) + (3/9)*math.log(9/3)

# Strong Entropy - 4N, 3Y
strongEnt = (4/7)*math.log(7/4) + (3/7)*math.log(7/3)


# Info Gain = 1 - AvgEntropy
windIG = 1 - (weakEnt*(9/16) + strongEnt*(7/16))
print("Wind Info Gain: ", windIG)


# Which Atrribute to choose at the root? We want the largest information gain
infoGains = [outlookIG, tempIG, humidityIG, windIG]
infoGains.sort()

print("The largest information gain is ", infoGains[-1])

print("The attribute chosen at the root is Humidity and Wind, since they have the same information gain.")

```
