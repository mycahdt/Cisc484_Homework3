```python
import numpy as np
import pandas
import matplotlib.pyplot as plt
import h5py
import scipy
import sklearn
```

```python
from sklearn.linear_model import LogisticRegression

from google.colab import drive
drive.mount('/content/drive')

pathTrainX = "/content/drive/MyDrive/3_Cisc484_MachineLearning/Homework3/mnist_train_data.npy"
pathTrainY = "/content/drive/MyDrive/3_Cisc484_MachineLearning/Homework3/mnist_train_labels.npy"
pathTestX = "/content/drive/MyDrive/3_Cisc484_MachineLearning/Homework3/mnist_test_data.npy"
pathTestY = "/content/drive/MyDrive/3_Cisc484_MachineLearning/Homework3/mnist_test_labels.npy"

# Load the MNIST dataset (assuming you have 'mnist_train_data.npy' and 'mnist_train_labels.npy' files)
x_train = np.load(pathTrainX)
y_train = np.load(pathTrainY)
x_test = np.load(pathTestX)
y_test = np.load(pathTestY)
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import SVC

# Use the one-vs-rest approach
logisticClassifier = LogisticRegression(solver='liblinear')
clf = OneVsRestClassifier(logisticClassifier).fit(x_train, y_train)
predicted = clf.predict(x_test)

print(predicted)
print(y_test)

if predicted.all() == y_test.all():
  print("They are the same")
```

```python
from sklearn import metrics as mt

# Test each of them on the test set and compute their individual accuracies.
for i in range(len(clf.estimators_)):
  pred = clf.estimators_[i].predict(x_test)
  print("1) Model ", i, " explains,", np.round(mt.explained_variance_score(y_test, pred)*100,2),"% variance of the target w.r.t features is")
  print("2) The Mean Absolute Error of model2 is:", np.round(mt.mean_absolute_error(y_test, pred),2))
  print("3) The R-Square score of the model2 is " , np.round(mt.r2_score(y_test,pred),2))
  print()

# The final accuracy
print("1) The model explains,", np.round(mt.explained_variance_score(y_test, predicted)*100,2),"% variance of the target w.r.t features is")
print("2) The Mean Absolute Error of model2 is:", np.round(mt.mean_absolute_error(y_test, predicted),2))
print("3) The R-Square score of the model2 is " , np.round(mt.r2_score(y_test,predicted),2))
```
